```python

import os
import json
from tabulate import tabulate
```

```python
def get_summary(root_dir, sort=True):
    tasks = []
    utility = []
    security = []
    for suite in os.listdir(root_dir):
        for task in os.listdir(os.path.join(root_dir, suite)):
            if task.startswith("user_task"):
                file = os.path.join(root_dir, suite, task, "none/none.json")
                with open(file, 'r') as f:
                    data = json.load(f)
                tasks.append(suite + "_" + data['user_task_id'])
                utility.append(data['utility'])
                security.append(data['security'])
    if sort:
        utility = [x[1] for x in sorted(zip(tasks, utility), key=lambda x: x[0])]
        security = [x[1] for x in sorted(zip(tasks, security), key=lambda x: x[0])]
        tasks = sorted(tasks)
    return tasks, utility, security

root_dir = "runs/meta-llama/Llama-3.3-70B-Instruct/local"
root_dir2 = "runs/meta-llama/Llama-3.3-70B-Instruct_dpo_NaiveCompletion_randpos_synthetic_alpaca_ns-1_bs512_1.6e-4_maxlength1024//local"

tasks, utility, security = get_summary(root_dir)
tasks2, utility2, security2 = get_summary(root_dir2)
```

```python
print(tabulate([tasks, utility, utility2], tablefmt="grid"))
```
